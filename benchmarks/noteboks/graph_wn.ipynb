```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open("../summary_wn.csv","r") as fd:
    lines=fd.readlines()

summary_uw=[]
summary_uw_header=[]
for line in lines:
    larr=line.strip().split(',')
    if len(larr)!=10:
        continue #bad line
    if larr[0][0]=='#':
        summary_uw_header=larr
    else:
        summary_uw.append(larr)
```

```python
i_n_samples=1
i_n_embs=2
i_org_cpu=3
i_cpu=4
i_gpu_v=6
i_gpu_3s=9
```

```python
org_cpu_n_samples=[]
cpu_n_samples=[]
gpu_v_n_samples=[]
gpu_3s_n_samples=[]
n_embs=[]
org_cpu=[]
cpu=[]
gpu_v=[]
gpu_3s=[]
for el in summary_uw:
    n_embs.append(int(el[i_n_embs]))
    if el[i_org_cpu]!='':
        org_cpu_n_samples.append(int(el[i_n_samples]))
        org_cpu.append(int(el[i_org_cpu]))
    if el[i_cpu]!='':
        cpu_n_samples.append(int(el[i_n_samples]))
        cpu.append(int(el[i_cpu]))
    if el[i_gpu_v]!='':
        gpu_v_n_samples.append(int(el[i_n_samples]))
        gpu_v.append(int(el[i_gpu_v]))
    if el[i_gpu_3s]!='':
        gpu_3s_n_samples.append(int(el[i_n_samples]))
        gpu_3s.append(int(el[i_gpu_3s]))
    
```

```python
fig, ax = plt.subplots()
#ax.plot(n_samples[:len(org_cpu)], org_cpu,
#        n_samples[:len(cpu)], cpu,
#        n_samples[:len(gpu_3s)], gpu_3s)
ax.plot(org_cpu_n_samples, org_cpu, label="org CPU Xeon Gold 6242")
ax.plot(cpu_n_samples, cpu, label="new CPU Xeon Gold 6242")
ax.plot(gpu_v_n_samples, gpu_v, label="new GPU V100")
ax.plot(gpu_3s_n_samples, gpu_3s, label="new GPU A40 fp32")

plt.title("Unifrac - weighted_normalized")
ax.legend(loc='upper left')
plt.show()
```
