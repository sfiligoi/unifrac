```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open("../mobile_summary_uw_f.csv","r") as fd:
    lines=fd.readlines()

summary_uw=[]
summary_uw_header=[]
for line in lines:
    larr=line.strip().split(',')
    if len(larr)!=8:
        continue #bad line
    if larr[0][0]=='#':
        summary_uw_header=larr
    else:
        summary_uw.append(larr)
```

```python
i_n_samples=1
i_n_embs=2
i_org_cpu=3
i_cpu=5
i_gpu_v=7
```

```python
org_cpu_n_samples=[]
cpu_n_samples=[]
gpu_v_n_samples=[]
n_embs=[]
org_cpu=[]
cpu=[]
gpu_v=[]
for el in summary_uw:
    n_samples=int(el[i_n_samples])
    if (n_samples%1000)!=0:
        continue # just the scaling
    n_embs.append(int(el[i_n_embs]))
    if el[i_org_cpu]!='':
        org_cpu_n_samples.append(int(el[i_n_samples]))
        org_cpu.append(int(el[i_org_cpu])/60.0)
    if el[i_cpu]!='':
        cpu_n_samples.append(int(el[i_n_samples]))
        cpu.append(int(el[i_cpu])/60.0)
    if el[i_gpu_v]!='':
        gpu_v_n_samples.append(int(el[i_n_samples]))
        gpu_v.append(int(el[i_gpu_v])/60.0)
    
```

```python
#x = n_samples
#y = np.vstack([org_cpu,cpu,gpu_3s])
```

```python
fig, ax = plt.subplots()
#ax.plot(n_samples[:len(org_cpu)], org_cpu,
#        n_samples[:len(cpu)], cpu,
#        n_samples[:len(gpu_3s)], gpu_3s)
ax.plot(org_cpu_n_samples, org_cpu, label="org CPU Mobile i7 8850H (MacOS)")
ax.plot(cpu_n_samples, cpu, label="new CPU Mobile i7 8850H fp32 (MacOS)")
ax.plot(gpu_v_n_samples, gpu_v, label="new GPU Mobile GTX 1050 fp32")

plt.ylim([0, 60])

plt.title("Unifrac -f (Laptop) - unweighted - in chip minutes")
ax.legend(loc='upper right')
plt.show()
```
